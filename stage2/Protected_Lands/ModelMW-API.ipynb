```python
from modelmw_client import *
```

mmw_secrets is just a .py file with:

`staging_api_key = [INSERT YOUR KEY]`

```python
from mmw_secrets import (
    staging_api_key,
)
```

```python
# Create an API user
mmw_run = ModelMyWatershedAPI(api_key=staging_api_key, use_staging=True)
```

```python
mmw_run.land_endpoint.format("2019_2019")
```

```python
import json
wcpa_json = json.load(open('wcpa-archive.geojson'))
```

```python
fd_json = json.load(open('fielddoc.geojson'))
```

```python
wcpa_json['features'][0]['geometry']
```

```python
lu_frame = mmw_run.run_batch_analysis(
    [wcpa_json['features'][0]['geometry'], fd_json['features'][0]['geometry']],
    mmw_run.land_endpoint.format("2019_2019"))
```
